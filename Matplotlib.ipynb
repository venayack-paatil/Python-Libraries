# **Matplotlib**

**Definition**

Matplotlib is a Python library used to create charts, graphs, and data visualizations.

It helps us understand data by representing it visually.

## **Why Matplotlib is Needed**
1.    Converts numerical data into visual graphs.
2.    Makes data easier to understand.
3.    Helps identify trends and patterns.
4.    Widely used in Data Science and Machine Learning.

# **Applications of Matplotlib**
*  Data Visualization
*  Machine Learning
*  Data Science
*  Business Reports
*  Stock Market Analysis
*  Healthcare Analytics
*  Scientific Research
*  Engineering Graphs

# **Import Matplotlib**

```python
# Import Matplotlib library
import matplotlib.pyplot as plt
```

# **Sample Data**

```python
# Create sample data

x = [1,2,3,4,5]
y = [10,20,30,40,50]
```

# **Line Plot**
**Function**

plt.plot()

```python
# Draw a line graph

plt.plot(x,y)

# Display the graph
plt.show()
```

# **Scatter Plot**
**Function**

plt.scatter()

```python
# Draw a scatter plot

plt.scatter(x,y)

plt.show()
```

# **Bar Chart**
**Function**

plt.bar()

```python
# Draw a bar chart

plt.bar(x,y)

plt.show()
```

# **Horizontal Bar Chart**
**Function**

plt.barh()


```python
# Draw a horizontal bar chart

plt.barh(x,y)

plt.show()
```

# **Histogram**
**Function**

plt.hist()

```python
# Sample data

marks=[60,70,70,80,85,90,95,95,100]

# Draw histogram

plt.hist(marks)

plt.show()
```

# **Pie Chart**
**Function**

plt.pie()

```python
# Student marks distribution

marks=[40,30,20,10]

subjects=["Math","Science","English","Python"]

plt.pie(marks,labels=subjects)

plt.show()
```



# **Box Plot**
**Function**

plt.boxplot()



```python
# Draw box plot

marks=[45,50,60,70,80,90,100]

plt.boxplot(marks)

plt.show()
```

# **Area Plot**
**Function**

plt.fill_between()

```python
# Draw area plot

plt.fill_between(x,y)

plt.show()
```

# **Titles & Labels**
**Functions**
* title()
* xlabel()
* ylabel()

```python
# Add title and axis labels

plt.plot(x,y)

plt.title("Student Marks")

plt.xlabel("Roll Number")

plt.ylabel("Marks")

plt.show()
```

# **Grid**
**Function**

grid()

```python
# Add grid lines

plt.plot(x,y)

plt.grid(True)

plt.show()
```

# **Legend**
**Function**

legend()

```python
# Display legend

plt.plot(x,y,label="Marks")

plt.legend()

plt.show()
```

# **Line Style**
**Function**

linestyle

```python
plt.plot(x,y,linestyle="--")

plt.show()
```

# **Line Color**
**Function**

color

```python
plt.plot(x,y,color="red")

plt.show()
```

# **Marker**
**Function**

marker

```python
plt.plot(x,y,marker="o")

plt.show()
```

# **Figure Size**
**Function**

figure()

```python
plt.figure(figsize=(8,5))

plt.plot(x,y)

plt.show()
```

# **Save Figure**
**Function**

savefig()

```python
plt.plot(x,y)

plt.savefig("graph.png")
```

# **Multiple Graphs**
**Function**

subplot()

```python
plt.subplot(1,2,1)

plt.plot(x,y)

plt.subplot(1,2,2)

plt.bar(x,y)

plt.show()
```
