# **Definition**

Pandas is an open-source Python library used for reading, organizing, cleaning, analyzing, and manipulating data.

It provides two main data structures:

Series (1D)                    

DataFrame (2D Table)

# **Why Pandas is Needed**

Python lists and dictionaries are good for storing small amounts of data, but they become difficult to manage when working with thousands or millions of rows.

**Pandas makes data handling:**



1.   Faster
2.   Easier


1.   Cleaner
2.   More efficient







# **Importance of Pandas**

*   Organizes data into rows and columns.

*   Reads data from CSV, Excel, JSON, SQL, etc.



*   Cleans missing or incorrect data.
*   Filters and sorts records.


*   Performs statistical analysis.
*  Prepares data for Machine Learning.

# **Import Pandas**

```python
# Import the Pandas library
import pandas as pd
```

# Creating **Data**

```python
# 1. Series
# Create a one-dimensional Series
s = pd.Series([10,20,30,40])
print(s)

# 2. DataFrame
# Create a table using a dictionary
data = {
    "Name":["Amit","Rahul","Priya"],
    "Age":[20,21,19]
}

df = pd.DataFrame(data)
print(df)

# 3. Read CSV
# Read a CSV file
# df = pd.read_csv("students.csv")
print(df)

# 4. Read Excel
# Read an Excel file
# df = pd.read_excel("students.xlsx")
print(df)

# 5. Read JSON
# Read a JSON file
# df = pd.read_json("students.json")
print(df)
```

# **Data-Frame Information**

```python
#head()
# Display first 5 rows
print(df.head())

# tail()
# Display last 5 rows
print(df.tail())

# info()
# Display complete information about the DataFrame
print(df.info())

# describe()
# Display statistical summary
print(df.describe())

# shape
# Display rows and columns
print(df.shape)

# columns
# Display all column names
print(df.columns)

# dtypes
# Display data type of every column
print(df.dtypes)

# index
# Display row index
print(df.index)
```

# **Selecting Data**

```python
# Select Name column
print(df["Name"])

# Select first row using index label
print(df.loc[0])

# Select first row using integer position
print(df.iloc[0])
```
