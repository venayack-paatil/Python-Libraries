## Cell 2 – Import Keras

```
# Import TensorFlow
import tensorflow as tf

# Import Keras
from tensorflow import keras

# Import NumPy
import numpy as np

# Import Matplotlib
import matplotlib.pyplot as plt
```

## Cell 3 – Purpose of Keras

### Definition
Keras is a high-level Deep Learning API built into TensorFlow.

### Why Keras?
- Easy to build Deep Learning models
- Beginner friendly
- Less code than raw TensorFlow
- Supports CNNs, RNNs, LSTMs and Transfer Learning

### Importance
- Simple syntax
- Fast development
- Industry standard with TensorFlow
- CPU/GPU/TPU support

### Applications
- Image Classification
- Face Recognition
- Chatbots
- Medical Diagnosis
- Recommendation Systems


## Cell 4 – Create Sample Dataset

```
X = np.array([[1],[2],[3],[4],[5]], dtype=np.float32)
y = np.array([[2],[4],[6],[8],[10]], dtype=np.float32)

print(X)
print(y)
```

## Cell 5 – Create Sequential Model

```
model = keras.Sequential([
    keras.Input(shape=(1,)),
    keras.layers.Dense(8, activation="relu"),
    keras.layers.Dense(1)
])

model.summary()
```

## Cell 6 – Add Dense Layers

```
model = keras.Sequential([
    keras.Input(shape=(1,)),
    keras.layers.Dense(16, activation="relu"),
    keras.layers.Dense(8, activation="relu"),
    keras.layers.Dense(1)
])

model.summary()
```

## Cell 7 – Compile Model

```
model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)
```

## Cell 8 – Train Model

```
history = model.fit(
    X,
    y,
    epochs=100,
    verbose=1
)
```

## Cell 9 – Training History

```
print(history.history)
```

## Cell 10 – Plot Loss Curve

```
plt.plot(history.history["loss"])
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

## Cell 11 – Predict

```
prediction = model.predict(np.array([[6]], dtype=np.float32))
print(prediction)
```

## Cell 12 – Evaluate Model

```
loss, mae = model.evaluate(X, y)
print("Loss:", loss)
print("MAE:", mae)
```

## Cell 13 – Save Model

```
model.save("student_marks.keras")
```

## Cell 14 – Load Model

```
loaded_model = keras.models.load_model("student_marks.keras")
print(loaded_model)
```

## Cell 15 – Model Summary

```
model.summary()
```

## Cell 16 – Dropout Layer

```
model = keras.Sequential([
    keras.Input(shape=(10,)),
    keras.layers.Dense(32, activation="relu"),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(1)
])

model.summary()
```

## Cell 17 – Flatten Layer

```
model = keras.Sequential([
    keras.Input(shape=(28,28)),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation="relu"),
    keras.layers.Dense(10, activation="softmax")
])

model.summary()
```

## Cell 18 – Early Stopping

```
callback = keras.callbacks.EarlyStopping(
    monitor="loss",
    patience=5
)
print(callback)
```

## Cell 19 – Model Checkpoint

```
checkpoint = keras.callbacks.ModelCheckpoint(
    "best_model.keras",
    save_best_only=True
)
print(checkpoint)
```

## Cell 20 – TensorBoard Callback

```
tensorboard = keras.callbacks.TensorBoard(log_dir="logs")
print(tensorboard)
```
