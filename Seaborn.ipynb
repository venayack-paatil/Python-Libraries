# **Seaborn**
**Definition**

Seaborn is a Python data visualization library built on top of Matplotlib.

It provides beautiful, informative, and statistical graphs with less code.

# **Why Seaborn is Needed**
*   Creates attractive statistical graphs.
*   Works directly with Pandas DataFrames.
*   Requires less code than Matplotlib.
*   Makes data easier to understand.

# **Importance of Seaborn**
* Better visualization than Matplotlib.
* Easy statistical plotting.
* Built-in themes and color palettes.
* Used in Data Science, AI, and * Machine Learning.

# **Applications of Seaborn**
*  Data Analysis
*  Data Visualization
*  Machine Learning
*  Exploratory Data Analysis (EDA)
*  Business Analytics
*  Financial Analysis
*  Healthcare Analytics
*  Research

# **Import Seaborn**

```python
# Import Seaborn library
import seaborn as sns

# Import Matplotlib for displaying graphs
import matplotlib.pyplot as plt

# Import Pandas
import pandas as pd
```

# **Create Sample Dataset**

```python
# Create sample dataset

data = {
    "Name":["Amit","Rahul","Priya","Sneha","Rohan"],
    "Age":[20,21,19,22,20],
    "Marks":[85,90,78,88,95],
    "Gender":["Male","Male","Female","Female","Male"]
}

df = pd.DataFrame(data)

print(df)
```

# **Line Plot**
**Function**

sns.lineplot()

```python
# Draw a line plot

sns.lineplot(data=df, x="Age", y="Marks")

plt.show()
```

# **Scatter Plot**
**Function**

sns.scatterplot()

```python
# Draw a scatter plot

sns.scatterplot(data=df, x="Age", y="Marks")

plt.show()
```

# **Bar Plot**
**Function**

sns.barplot()

```python
# Display average marks by gender

sns.barplot(data=df, x="Gender", y="Marks")

plt.show()
```

# **Histogram**
**Function**

sns.histplot()

```python
# Display the distribution of marks

sns.histplot(df["Marks"])

plt.show()
```

# **KDE Plot**
**Function**

sns.kdeplot()

```python
# Display smooth distribution curve

sns.kdeplot(df["Marks"])

plt.show()
```

# **Box Plot**
**Function**

sns.boxplot()

```python
# Display spread of marks

sns.boxplot(data=df, y="Marks")

plt.show()
```

# **Violin Plot**
**Function**

sns.violinplot()

```python
# Display data distribution

sns.violinplot(data=df, y="Marks")

plt.show()
```

# **Count Plot**
**Function **

sns.countplot()

```python
# Count Male and Female students

sns.countplot(data=df, x="Gender")

plt.show()
```

# **Heatmap**
**Function**

sns.heatmap()

```python
# Calculate correlation

corr = df[["Age","Marks"]].corr()

# Display heatmap

sns.heatmap(corr, annot=True)

plt.show()
```

# **Pair Plot**
**Function**

sns.pairplot()

```python
# Display relationship between numerical columns

sns.pairplot(df)

plt.show()
```

# **Pair Plot with Hue**

```python
# Display relationship by gender

sns.pairplot(df, hue="Gender")

plt.show()
```

# **Joint Plot**
**Function**

sns.jointplot()

```python
# Relationship between Age and Marks

sns.jointplot(data=df, x="Age", y="Marks")

plt.show()
```

# **Regression Plot**
**Function**

sns.regplot()

```python
# Display regression line

sns.regplot(data=df, x="Age", y="Marks")

plt.show()
```

# **Strip Plot**
**Function**

sns.stripplot()

```python
# Display individual values

sns.stripplot(data=df, x="Gender", y="Marks")

plt.show()
```

# **Swarm Plot**
**Function**

sns.swarmplot()

```python
# Display non-overlapping data points

sns.swarmplot(data=df, x="Gender", y="Marks")

plt.show()
```

# **Themes**
**Function**

sns.set_theme()

```python
# Apply a dark grid theme

sns.set_theme(style="darkgrid")

sns.barplot(data=df, x="Gender", y="Marks")

plt.show()
```

# **Color Palette**
**Function**

sns.color_palette()

```python
# Change graph colors

sns.barplot(
    data=df,
    x="Gender",
    y="Marks",
    palette="Set2"
)

plt.show()
```
