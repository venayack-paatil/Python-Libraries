# **Create Tensor**
**Function**

tf.constant()

```python
# Create a constant tensor

tensor = tf.constant([10,20,30,40])

print(tensor)
```

# **Create Variables**
**Function**

tf.Variable()

```python
# Create a variable tensor

variable = tf.Variable([1,2,3])

print(variable)
```

# **Tensor Shape**
**Function**

tf.shape()

```python
# Display tensor shape

tensor = tf.constant([[1,2],[3,4]])

print(tf.shape(tensor))
```

# **Tensor Operations**
**Functions**

tf.add()

tf.subtract()

tf.multiply()

tf.divide()

```python
# Create tensors

a = tf.constant([10,20])

b = tf.constant([5,2])

print(tf.add(a,b))

print(tf.subtract(a,b))

print(tf.multiply(a,b))

print(tf.divide(a,b))
```

# **Random Tensor**

tf.random.normal()

# **Sequential Model**
**Function**

tf.keras.Sequential()

```python
# Create neural network

import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(8, activation="relu"),
    tf.keras.layers.Dense(1)
])

print(model)
```

# **Compile Model**
**Function**

compile()

```python
model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)
```

# **Train Model**
**Function**

fit()

```python
import numpy as np

X = np.array([[1], [2], [3], [4], [5]], dtype=np.float32)
y = np.array([[2], [4], [6], [8], [10]], dtype=np.float32)

model.fit(X, y, epochs=50)
```

# **Predict**
**Function**

predict()

```python
# Predict output

prediction = model.predict(np.array([6]))

print(prediction)
```

# **Evaluate Model**
**Function**

evaluate()

```python
# Evaluate model

model.evaluate(X,y)
```

# **Save Model**
**Function**

save()

```python
# Save trained model

model.save("model.keras")
```

# **Load Model**
**Function**

load_model()

```python
# Load saved model

loaded_model = tf.keras.models.load_model("model.keras")

print(loaded_model)
```

# **Dense Layer**
**Function**

Dense()

```python
# Create Dense layer

layer = tf.keras.layers.Dense(
    units=64,
    activation="relu"
)

print(layer)
```

# **Dropout Layer**
**Function**

Dropout()

```python
# Create Dropout layer

layer = tf.keras.layers.Dropout(0.3)

print(layer)
```

# **Flatten Layer**
**Function**

Flatten()

```python
# Flatten image

layer = tf.keras.layers.Flatten()

print(layer)
```

# **Convolution Layer**
**Function**

Conv2D()

```python
# Create convolution layer

layer = tf.keras.layers.Conv2D(
    filters=32,
    kernel_size=(3,3),
    activation="relu"
)

print(layer)
```

# **Max Pooling**
**Function**

MaxPooling2D()

```python
# Create pooling layer

layer = tf.keras.layers.MaxPooling2D(
    pool_size=(2,2)
)

print(layer)
```

# **TensorBoard**
**Function**

TensorBoard()

```python
# Create TensorBoard callback

callback = tf.keras.callbacks.TensorBoard(
    log_dir="logs"
)

print(callback)
```
